```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
x1 = np.array([1,2,3,4,5,6])
x2 = np.array([7.5,11,16,18,20,26])
```

```python
plt.scatter(x1,x2)
plt.show()
```

```python
X=np.c_[x1,x2]
X
```

```python
pca = PCA(n_components = 1)
X_transformed = pca.fit_transform(X)
X_transformed
```

```python
pca.components_      # it will tell the direction . first is ist quadrant and 2 will be in 4 quadrant .here first one is imp..
```

```python
pca.inverse_transform(X_transformed)
```

```python
X_aprox = pca.inverse_transform(X_transformed)
```

```python
plt.scatter(X_aprox[:,0],X_aprox[:,1])
plt.show()
```
